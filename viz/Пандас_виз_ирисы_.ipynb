```python
import pandas as pd
from sklearn.datasets import load_iris

# Загрузка Iris-датасета
iris = load_iris()
# Создание DataFrame из данных датасета
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Добавление столбца с метками классов
df['class'] = iris.target

# Вывод первых нескольких строк датасета
df.head()
```

```python
df.plot()
```

```python
df["petal length (cm)"].plot(figsize=(15,3))
```

```python
df.groupby("class")["petal length (cm)"].sum().plot(figsize=(15,3))
```

```python
df.groupby("class")["petal length (cm)"].sum().plot(kind='bar',figsize=(15,3))
```

```python
df.groupby("class")["petal length (cm)"].sum().plot(kind='barh',figsize=(15,3))
```

```python
df["sepal length (cm)"].plot.hist(figsize=(15,3))
```

```python
pd.plotting.scatter_matrix(df);
```

```python
pd.plotting.scatter_matrix(df[["sepal length (cm)",
                               "sepal width (cm)"]], figsize=(15,7));
```

```python
pd.plotting.scatter_matrix(df[["sepal length (cm)",
                               "sepal width (cm)"]],
                           figsize=(15,7),
                           diagonal='kde',
                           color='red');
```

```python
bins = [0,5,10]
labels = ["'0-5'","'6-10'"]
df['sepal_length_bucket'] = pd.cut(df['sepal length (cm)'],
                                   bins=bins, labels=labels)
df.head()
```

```python
pd.pivot_table(data=df, index = ["class"],
               columns=["sepal_length_bucket"],
               values=["sepal width (cm)"])
```

```python
pd.pivot_table(data=df, index = ["class"],
               columns=["sepal_length_bucket"],
               values=["sepal width (cm)"]).plot.barh(figsize=(15,3))
```
