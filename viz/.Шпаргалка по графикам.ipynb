```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
olympic = "https://github.com/ernest-kiwele/120-years-of-olympic-history--athletes-and-results/blob/master/data/athlete_events.csv?raw=true"
olympic = pd.read_csv(olympic)

iris="https://github.com/Vitalyastiy/courses/blob/main/kaggle/iris.csv?raw=true"
iris = pd.read_csv(iris)
```

```python
# пример полиномиальной кривой
fig, ax = plt.subplots(figsize= (14,4))
ax.scatter(x =iris['Sepal Width (cm)'], y =iris['Petal Length (cm)'] )
x =iris['Sepal Width (cm)']
y =iris['Petal Length (cm)']
z = np.polyfit(x, y, 3)
p = np.poly1d(z)

ax.plot(x, p(x), color = 'red')

ax.set_xlabel('Height')
ax.set_ylabel('Weight')
plt.show()
```

```python
#категоирийный скаттерплот
sns.set_style("dark") #darkgrid, whitegrid, dark, white, ticks
sns.scatterplot(data = iris, x = 'Sepal Length (cm)', y = 'Sepal Width (cm)', hue = 'Species') # , marker = '*'
plt.title("Sepal Length (cm) vs. Sepal Width (cm)")
plt.xlabel('Длинна чашелистика')
```

```python
#линия регрессии
sns.regplot(data = iris, x = 'Sepal Length (cm)', y = 'Sepal Width (cm)')
```

```python
# использование палитры
sns.scatterplot(x='Age', y='Height', hue='Sex', data=olympic, palette='Set2') # "deep" "muted" "bright" "pastel" "dark" "colorblind"
plt.title('Age vs. Height')
```

```python
#linelot категорийныйN
sns.lineplot(data=olympic, x = 'Age', y = 'Height', hue = 'Sex' , estimator = 'median')
```

```python
#гистограмма по бинам
count, bin_edges = np.histogram(iris['Sepal Length (cm)']) #вычисляет количество элементов в каждом бине (count) и границы бинов (bin_edges)
iris['Sepal Length (cm)'].plot(kind = "hist", bins = 50, figsize=(12,4), xticks=bin_edges )
```

```python
# гистограмма + ядерная оценка плотности
f, ax = plt.subplots(figsize=(7, 5))
sns.despine(f)
sns.distplot(olympic['Age'], kde=False)
```

```python
# ядерная оценка плотности/гистограмма
f, ax = plt.subplots(figsize=(7, 5))
sns.despine(f)
#sns.distplot(iris['Petal Length (cm)'], kde=True)
sns.kdeplot(iris['Petal Length (cm)'], shade = True , vertical=True)
```

```python
# совмещение KDE на одно графике
f, ax = plt.subplots(figsize=(7, 5)) # , nrows= 1 , ncols= 3
#ig, axs = plt.subplots(nrows= 1 , ncols= 3)
sns.despine(f)

sns.kdeplot(iris['Petal Length (cm)'], shade=True, color = 'm')
sns.kdeplot(iris['Petal Width (cm)'], shade=True, color = 'g')

plt.show()
```
